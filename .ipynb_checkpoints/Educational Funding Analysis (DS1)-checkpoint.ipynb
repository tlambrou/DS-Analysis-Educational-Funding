```python
# %load /Users/Tassos/Desktop/Academics/Make-School/Product-College/DS1\ -\ Data\ Analysis/Course\ Work/DS-Analysis-Educational-Funding/dataset/us-educational-finances/chew_data.py
"""
A script for compressing U.S. Census financial data on schools into a single file.
"""

import re
import zipfile
import os
import pandas as pd
import us

# The name of the output CSV
OUTPUT_NAME = 'elsect_main.csv'

# The name of the zip file being unpacked
ZIP_NAME = 'elsect.zip'

# A hard-coded schema that matches the column structure
# Can be modified to include additional columns
SCHEMA = ['STATE',
          'ENROLL',
          'NAME',
          'YRDATA',
          'TOTALREV',
          'TFEDREV',
          'TSTREV',
          'TLOCREV',
          'TOTALEXP',
          'TCURINST',
          'TCURSSVC',
          'TCURONON',
          'TCAPOUT']


STATES = us.STATES

# Useful regular expressions
doublespace = re.compile(r'  ')
numbersonly = re.compile(r'\d+')


def spreadsheet_to_dataframe(filename):
    """
    Converts a spreadsheet to a Pandas dataframe.
    Performs minor alterations to the data
    (i.e. changing '1' to 'Alabama').

    :param filename: The name of the xls file.
    :return: A dictionary with the U.S. states as keys.
    :rtype: dict
    """

    # Give the user a heads-up
    print('Parsing ' + str(filename) + '...')

    # Convert the year data to full years
    # Can't depend on spreadsheet value, so we'll do it by filename
    year = int(re.findall(numbersonly, filename)[0])
    if year < 10:
        year = '200' + str(year)
    elif year < 50:
        year = '20' + str(year)
    else:
        year = '19' + str(year)

    # Use year to determine state identifier
    # (column name changes in 2002)
    if int(year) >= 2002:
        st_code = ['IDCENSUS']
    elif int(year) == 1992:
        st_code = ['ID']
    else:
        st_code = ['GOVSID']
    specific_schema = st_code + SCHEMA

    # Open the file, parse it, and truncate the highlights
    data = pd.read_excel(filename, sheet_name=0, dtype=str)
    data = pd.DataFrame(data, columns=specific_schema)

    # Convert the state codes to state names
    data[st_code] = data[st_code].applymap(lambda x: x[:2])
    data['STATE'] = data[st_code].applymap(lambda x: us.STATES[int(x) - 1])
    data = data.drop(st_code, axis=1)

    # Fill in the year data
    data['YRDATA'] = year

    return data


# Unpack the data
out = zipfile.ZipFile(ZIP_NAME, 'r')
file_list = out.namelist()
file_list.remove('elsect/')
out.extractall(os.getcwd())
out.close()

# Iterate through spreadsheets, extracting data
record = []
for item in file_list:
    df = spreadsheet_to_dataframe(item)
    record.append(df)

# Glue the annual surveys into a single file
output = pd.concat(record)

# Write to file as CSV
output.to_csv(OUTPUT_NAME, index=False)

# Clean up
for item in file_list:
    os.remove(item)

print('')
print('Finished.')



```

```python
# %load /Users/Tassos/Desktop/Academics/Make-School/Product-College/DS1\ -\ Data\ Analysis/Course\ Work/DS-Analysis-Educational-Funding/dataset/us-educational-finances/state_summary.py
"""
A script for creating a state summary file from the education financial data.
Should be run after "chew_data.py".
"""

import pandas as pd
import sqlite3

INPUT_FILENAME = 'elsect_main.csv'
OUTPUT_FILENAME = 'elsect_summary.csv'

SCHEMA = ['STATE',
          'ENROLL',
          'NAME',
          'YRDATA',
          'TOTALREV',
          'TFEDREV',
          'TSTREV',
          'TLOCREV',
          'TOTALEXP',
          'TCURINST',
          'TCURSSVC',
          'TCURONON',
          'TCAPOUT']

query = \
    '''
    SELECT STATE AS State,
    YRDATA AS Year,
    SUM(ENROLL) AS Enrollment,
    SUM(TOTALREV) AS Total_Revenue,
    SUM(TFEDREV) AS Federal_Revenue,
    SUM(TSTREV) AS State_Revenue,
    SUM(TLOCREV) AS Local_Revenue,
    SUM(TOTALEXP) AS Total_Expenditure,
    SUM(TCURINST) AS Instruction_Expenditure,
    SUM(TCURSSVC) AS Support_Services_Expenditure,
    SUM(TCURONON) AS Other_Expenditure,
    SUM(TCAPOUT) AS Capital_Outlay_Expenditure
    FROM school_money
    GROUP BY STATE, YRDATA
    ORDER BY YRDATA;
    '''

# Create a temporary SQL database populated with school data
con = sqlite3.connect(':memory:')
cur = con.cursor()
df = pd.read_csv(INPUT_FILENAME)
df.to_sql(name='school_money', con=con, if_exists='replace')
con.commit()

# Run a query to aggregate data by state
output = pd.read_sql(sql=query, con=con)
output.to_csv(OUTPUT_FILENAME, index=False)

print('Finished.')
```

```python
def multi_distplot(df, columns, log=False):
    f, ax = plt.subplots(3,3,figsize=(12, 12), sharey=False)
    for row,data in enumerate([columns[0:3],columns[3:6],columns[6:9]]):
        for i,y in enumerate(data):
            sns.kdeplot(summary_df[y],shade=True,ax=ax[row,i])
            if log is True:
                ax[row,i].set_xscale('log')
            ax[row,i].set_title("Distribution of {}".format(y))
            ax[row,0].set_ylabel("Density")
            ax[row,i].set_xlabel("USD in Log Scale")
    plt.tight_layout(pad=0)
```

```python
multi_distplot(df=summary_df, columns=summary_df.columns[2:11], log=True)
plt.show()
```

```python
def state_multi_distplot(df, statevar, columns, log=False):
    f, ax = plt.subplots(3,3,figsize=(12, 12), sharey=False)
    for row,data in enumerate([columns[0:3],columns[3:6],columns[6:9]]):
        for i,y in enumerate(data):
            for state in summary_df[statevar][summary_df[statevar].notnull()].unique():
                sns.kdeplot(summary_df[y][summary_df[statevar] == state],shade=False,ax=ax[row,i], label=state)
            if log is True:
                ax[row,i].set_xscale('log')
            ax[row,i].set_title("Distribution of {}".format(y))
            ax[row,0].set_ylabel("Density")
            ax[row,i].set_xlabel("USD in Log Scale")
            ax[row,i].legend_.remove()
    plt.tight_layout(pad=0)
```

```python
state_multi_distplot(df=summary_df, statevar="State", columns=summary_df.columns[2:11], log=True)
plt.show()
```

```python
# State List
states_list = summary_df["State"][summary_df["State"].notnull()].unique()

# Aggregate the States for General Time-Series
df_mean = (summary_df.mean()/1000).reset_index()
df_mean.columns =["Variable","Mean"]

# Since these variables operate at a different level of magnitude, split them.
highmag = list(df_mean["Variable"][(df_mean["Mean"] > 2000) & (df_mean["Variable"] != "Year")])
lowmag = [x for x in list(df_mean["Variable"][df_mean["Variable"] != "Year"]) if x not in highmag]
print("High Magnitude Variables:\n",highmag)
print("\nLow Magnitude Variables:\n",lowmag)
```

```python
f, ax = plt.subplots(1,2, figsize=[11,4])
for i,x in enumerate([lowmag,highmag]):
    (summary_df[x+["Year"]].groupby(["Year"]).mean()/1000).plot(ax=ax[i])
    #ax[i].legend(fontsize='large', loc='center left',bbox_to_anchor=(1, 0.5))
    ax[i].set_title("Aggregated Revenue and Expenditure over Time")
    ax[i].set_xlabel("Year")
ax[0].set_ylabel("USD in Thousands")
plt.tight_layout(pad=0)
plt.show()
```

```python
# State/Year Aggregate
def aggr_year_state(summary_df, var, n=7):
    temp = (summary_df[[var,"Year","State"]].groupby(["Year", "State"]).mean()/1000).reset_index().set_index("Year")
    temp = temp[temp.notnull()]
    volatile = temp.groupby("State").std().sort_values(by=var, ascending=False).reset_index().loc[:n,"State"]
    return temp, volatile

# Volatility Plot
def volatile_subplots(summary_df, variables):
    f, ax = plt.subplots(len(variables), figsize=[10,9], sharex=True)
    for i,var in enumerate(variables):
        temp, volatile= aggr_year_state(summary_df,var)
        for state in volatile:
            ax[i].plot(temp[var][temp["State"] == state].diff(), label=state)
        ax[i].legend(fontsize='large', loc='center left',bbox_to_anchor=(1, 0.5))
        ax[i].set_title("Yearly Difference for {} - Top 7 Volatile States".format(var))

        ax[i].set_ylabel("Difference in Thousands USD")
    ax[len(variables)-1].set_xlabel("Year")
    plt.tight_layout(pad=0)
```

```python
volatile_subplots(summary_df=summary_df, variables= ["Total_Revenue", "Total_Expenditure", "Enrollment"])
plt.show()
```

```python
summary_df["Rev-Exp"] =  (summary_df["Total_Revenue"] - summary_df["Total_Expenditure"])/summary_df["Total_Revenue"]
```

### This metric is built to quantify how big the gap is between Revenue and Expenditure, while normalizing the value to compare it between states.

```python
sns.distplot(summary_df["Rev-Exp"])
plt.show()
```

```python
def state_year_plot(df, var,states):
    f, ax = plt.subplots(1, figsize=[8,5])
    for state in states:
        ax.plot(df[var][df["State"] == state], label=state)
    ax.legend(fontsize='large', loc='center left',bbox_to_anchor=(1, 0.5))
    ax.set_xlabel("Year")
    ax.set_ylabel("Difference Scaled by Total Revenue")
    ax.set_title("Difference Between Total Revenue and Expenditure by State")
```

```python
# New DF
benefit = round(summary_df[["Rev-Exp","Year","State"]].groupby(["Year","State"]).mean().mul(100),2).reset_index().set_index("Year")
# Top / Bottom States
sort_exp = benefit.groupby("State").mean().sort_values(by="Rev-Exp", ascending=False).reset_index()
top = list(sort_exp["State"].head())
bot = list(sort_exp["State"].tail())
# Plot
state_year_plot(df=benefit, var="Rev-Exp", states= top+bot)
```
